```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}"

message = greet("Alice")  # 不传递第二个参数，使用默认值
print(message)  # 输出: Hello, Alice

message = greet("Bob", "Hi")  # 传递第二个参数，覆盖默认值
print(message)  # 输出: Hi, Bob
```
